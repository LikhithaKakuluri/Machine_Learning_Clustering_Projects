# Rescaling

```python
from sklearn.preprocessing import StandardScaler

df1= df[["Age","Annual Income (k$)","Spending Score (1-100)"]]

# Class instance
scaler = StandardScaler()

# Fit_transform
data=scaler.fit_transform(df1)

data= pd.DataFrame(data)
data.columns = ["Age","Annual Income (k$)","Spending Score (1-100)"]
data.head()
```

```python
x= data[["Annual Income (k$)","Spending Score (1-100)"]]
x
```

# Model Building

```python
kmeans = KMeans(n_clusters=4, max_iter=50,n_init=10)
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# Calculating sum of squared distance

ssd = []

for num_clusters in range(1,11):
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50,n_init=10)
    kmeans.fit(x)
    
    ssd.append(kmeans.inertia_)
print("Calculated Sum of Square distance values are:")
ssd
```

# Finding the Optimal Number of curves

#PLOTTING A ELBOW GRAPH

The elbow method is a graphical representation of finding the optimal 'K' in a K-means clustering. It works by finding WCSS (Within-Cluster Sum of Square) i.e. the sum of the square distance between points in a cluster and the cluster centroid.

```python
sns.set()
plt.plot(range(1,11),ssd,color='red',marker="*")
plt.title("The Elbow Point Graph")
plt.xlabel("Number of Clusters")
plt.ylabel('WCSS')
plt.show()
```

Result :
    
From the above Elbow curve we will take the value of K=5 and rebuild the cluster.

```python
# K=5

kmeans = KMeans(n_clusters=5, max_iter=50)
kmeans.fit(x)
kmeans.labels_

df["Label"] = kmeans.labels_

df.head()
```

```python
plt.figure(figsize=(10,6))

plt.title("Ploting the data into 5 clusters",fontsize=20, color="green")
sns.scatterplot(data=df, x="Annual Income (k$)", y="Spending Score (1-100)", hue="Label", s=60, palette=['green','orange','brown','blue','red'])
plt.show()
```

Result :

    Data is plotted into 5 clusters
